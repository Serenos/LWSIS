```python
original_images = [x["image"].to(device) for x in batched_inputs]
gt_instances = [x["instances"].to(device) for x in batched_inputs]

original_image_masks = [torch.ones_like(x[0], dtype=torch.float32) for x in original_images]

from detectron2.modeling import build_model
cfg.MODEL.DEVICE = 'cpu'
model = build_model(cfg)

original_images = ImageList.from_tensors(original_images, model.backbone.size_divisibility)
original_image_masks = ImageList.from_tensors(
    original_image_masks, model.backbone.size_divisibility, pad_value=0.0
)

model.add_bitmasks_from_boxes(gt_instances, original_images.tensor, original_image_masks.tensor, original_images.tensor.size(-2), original_images.tensor.size(-1))



gt_instances

gt_instances[0].image_color_similarity.size()

image_color_similarity = torch.cat([x.image_color_similarity for x in gt_instances])

image_color_similarity.size()

pairwise_color_thresh = 0.3

weigths = (image_color_similarity >= pairwise_color_thresh)

weigths.size()

edge_num = weigths.size(0)*weigths.size(1)*weigths.size(2)*weigths.size(3)

edge_num

weigths.float().sum()

prop_pos_edges = []

import numpy as np

pairwise_color_threshs = np.linspace(0.0,1.0,11)

for i in pairwise_color_threshs:
    weigths = (image_color_similarity >= i)
    prop_pos_edges.append(weigths.float().sum() / edge_num)
```

```python
plt.plot(pairwise_color_threshs, prop_pos_edges)
plt.show()
```
